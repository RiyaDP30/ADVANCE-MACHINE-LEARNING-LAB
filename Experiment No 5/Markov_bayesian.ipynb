```python
# Define states and actions
states = [0, 1, 2]
actions = ['a', 'b']

# Transition probabilities P[s][a][s']
P = {
    0: {'a': [0.5, 0.5, 0.0], 'b': [0.2, 0.8, 0.0]},
    1: {'a': [0.0, 0.6, 0.4], 'b': [0.1, 0.0, 0.9]},
    2: {'a': [0.0, 0.0, 1.0], 'b': [0.0, 0.0, 1.0]}
}

# Rewards R[s][a]
R = {
    0: {'a': 5, 'b': 3},
    1: {'a': 4, 'b': 6},
    2: {'a': 0, 'b': 0}
}

gamma = 0.9   # Discount factor
theta = 0.001 # Convergence threshold

# Initialize value function
V = np.zeros(len(states))

def value_iteration():
    global V
    while True:
        delta = 0
        for s in states:
            v = V[s]
            action_values = []
            
            for a in actions:
                val = R[s][a]
                for s_next, prob in enumerate(P[s][a]):
                    val += gamma * prob * V[s_next]
                action_values.append(val)
            
            V[s] = max(action_values)
            delta = max(delta, abs(v - V[s]))
        
        if delta < theta:
            break

value_iteration()

print("Optimal Value Function:")
print(V)

# Derive policy
policy = {}
for s in states:
    action_values = {}
    for a in actions:
        val = R[s][a]
        for s_next, prob in enumerate(P[s][a]):
            val += gamma * prob * V[s_next]
        action_values[a] = val
    
    policy[s] = max(action_values, key=action_values.get)

print("\nOptimal Policy:")
print(policy)
```

```python
import numpy as np
import pandas as pd
```

```python
# Sample dataset
data = {
    'Feature1': [1, 2, 1, 2, 1, 2],
    'Feature2': [0, 1, 1, 0, 0, 1],
    'Class': ['Yes', 'Yes', 'No', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)

# Separate features and labels
X = df[['Feature1', 'Feature2']]
y = df['Class']

# Calculate prior probabilities
priors = y.value_counts(normalize=True)
print("Priors:\n", priors)

# Likelihood calculation
likelihood = {}

for feature in X.columns:
    likelihood[feature] = {}
    for value in X[feature].unique():
        likelihood[feature][value] = {}
        for cls in y.unique():
            count = len(df[(df[feature] == value) & (df['Class'] == cls)])
            total = len(df[df['Class'] == cls])
            likelihood[feature][value][cls] = count / total

print("\nLikelihoods:")
print(likelihood)

# Prediction function
def predict(sample):
    posteriors = {}
    
    for cls in priors.index:
        prob = priors[cls]
        
        for feature, value in sample.items():
            prob *= likelihood[feature][value][cls]
        
        posteriors[cls] = prob
    
    return max(posteriors, key=posteriors.get)

# Test sample
sample = {'Feature1': 1, 'Feature2': 0}
prediction = predict(sample)

print("\nSample:", sample)
print("Predicted Class:", prediction)
```
